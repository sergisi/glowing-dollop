```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pk = 'id'
msg = 'messages'
sign = 'signatures'
anonymity = 'anonymity'

def load(filep: str='data.csv'):
    data = pd.read_csv(filep, names=[pk, msg, sign], index_col=pk)
    data[anonymity] = data[sign] / data[msg]
    return data
data = load()
```
